```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from statsmodels.stats import weightstats as stests
import statsmodels.api as sm
```

```python
df = pd.read_csv('zomato.csv')

df.head()
```

```python
print 'Columns: ', df.shape[0]
print 'Rows: ', df.shape[1]
```

```python
#Dropping unnecessary columns from dataset.
to_drop = ['url','phone','menu_item']
df.drop(to_drop,inplace=True,axis=1)
df.head()
```

```python
df['address'].is_unique
```

```python
df.loc[:,'approx_cost(for two people)'].head(10)
```

```python
#Converting cost to numeric.
extr = df['approx_cost(for two people)'].str.replace(",","");

df['approx_cost(for two people)']=pd.to_numeric(extr)

df.head()
```

```python
df.loc[:,'rate'].head(10)
#Removing all new and unrated restaurants
df =df[df.rate!='NEW']
df = df[df.rate!='-']
df =df.dropna(axis=0,how='any')
```

```python
extr = df['rate'].str.slice(0,3)
extr
df['rate'] = pd.to_numeric(extr)
```

```python
# https://www.livemint.com/news/india/the-most-popular-dine-out-hubs-in-india-11570968468566.html
print("According to LiveMint, the average review for restaurants in Bangalore is <=4.1. The sample of restaurants taken was 1000")
print("Thus, H1 = Average review of restaurants in bangalore is > 4.1")
print("      H0 = Average review of restaurants in bangalore is <=4.1")
avg_reviews = 4.1 #up to 4.1
n = 1000
pnull = avg_reviews/100
sample_data = df.sample(n=1000)
phat = df['rate'].mean()/100
ztest,pval = sm.stats.proportions_ztest(phat * n, n, pnull, alternative='larger')
# ztest ,pval = stests.ztest(df['rate'], x2=None, value=avg_reviews/100)
# print(float(pval))
if pval<0.05:
     print("reject null hypothesis")
else:
     print("Accept null hypothesis. The average review of restaurants is <=4.1")
```

```python
# https://www.numbeo.com/cost-of-living/in/Bangalore
print("According to Numbeo, the average cost for a two person meal at a mid-tier restaurant in Bangalore is <=1,000. The sample of restaurants taken was 8154")
print("Thus, H1 = Average cost of a two-person meal at restaurants in bangalore is > 1000")
print("      H0 = Average cost of a two-person meal at restaurants in bangalore is <= 1000")
avg_reviews = 1000 #up to 4.1
n = 8154
pnull = avg_reviews/100
sample_data = df.sample(n=8154)
phat = df['approx_cost(for two people)'].mean()/100
ztest,pval = sm.stats.proportions_ztest(phat * n, n, pnull, alternative='larger')
# ztest ,pval = stests.ztest(df['rate'], x2=None, value=avg_reviews/100)
# print(float(pval))
if pval<0.05:
     print("reject null hypothesis")
else:
     print("Accept null hypothesis. The average cost of a two-person meal at restaurants in bangalore is <= 1000")
```
