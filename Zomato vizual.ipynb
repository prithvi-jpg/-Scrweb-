```python
#Pie,bar chart for area wise restaurant density

dict1={}
for i in df['listed_in(city)']:
    if i not in dict1:
        dict1[i]=0;
    dict1[i]+=1;

    
dict1
```

```python
restcount=[]
for x in dict1:
    restcount.append(dict1[x])
citycountdf = pd.DataFrame(list(dict1))
citycountdf['count']=restcount

citycountdf
```

```python



ax = citycountdf.plot(kind='bar',figsize=(18,10))
ax.set_title('Restaurant Frequency')
ax.set_xlabel('Restaurant Locations')
ax.set_ylabel('Frequency')
ax.set_xticklabels(citycountdf[0])
```

```python

plt.pie(citycountdf['count'],radius=4, labels=citycountdf[0],autopct='%1.3f',pctdistance=0.8)

plt.show()
```

```python
df['cuisines'].unique()
l_set=set()
l_list=[]
for i in df['cuisines']:
    l = i.split(',')
    for t in l:
        l_set.add(t.strip())
        l_list.append(t.strip())
print(len(l_set))
NI = l_list.count('North Indian')

import collections
counter=collections.Counter(l_list)
counter
```

```python
top_10 = counter.most_common(10)
m_df = pd.DataFrame(top_10,columns=['Cuisine','Count'])
print(m_df)
ax =m_df.plot(kind='bar',figsize=(18,10))
ax.set_title('Cuisine Frequency')
ax.set_xlabel('Cuisine Type')
ax.set_ylabel('Frequency')
ax.set_xticklabels(m_df['Cuisine'])
```
